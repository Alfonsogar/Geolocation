## Importar Tablas

### Regiones

```python
regiones = pd.read_table('./Microdato_Censo2017-Regiones.csv', sep=';')
regiones.columns = ['REGION', 'NOMBRE_REGION']
regiones.set_index('REGION', inplace=True)
```

```python
ds = ds.set_index('REGION').join(regiones,how='left', lsuffix='main')
```

### Provincias

```python
provincias = pd.read_table('./Microdato_Censo2017-Provincias.csv', sep=';')
provincias.set_index('PROVINCIA', inplace=True)
provincias.head()
```

```python
ds = ds.set_index('PROVINCIA').join(provincias, how='left', lsuffix='main')
```

### Comunas

```python
comunas = pd.read_table('./Microdato_Censo2017-Comunas.csv', sep=';')
comunas.set_index('COMUNA', inplace=True)
comunas.head()
```

```python
ds = ds.set_index('COMUNA').join(comunas, how='left', lsuffix='main')
```

### Area

```python
area = pd.read_table('./Microdato_Censo2017-Areas.csv', sep=';')
area.set_index('AREA', inplace=True)
area
```

```python
ds = ds.set_index('AREA').join(area, how='left')
```

```python
ds.to_csv('Output_censo.csv')
```
